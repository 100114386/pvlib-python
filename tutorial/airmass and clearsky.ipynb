````python
import pvlib
````

Make a location object

````python
tus = pvlib.Location(32.2, -111, 'US/Arizona', 700, 'Tucson')
print(tus)
````

Make a pandas.DatetimeIndex with the times we're interested in

````python
start = datetime.datetime(2014,6,23)
end = datetime.datetime(2014,6,24)
times = pd.date_range(start=start, end=end, freq='1Min')
````

Calculate the solar position at the times of interest. We could use ``solarposition.spa()`` or ``solarposition.pyephem()``, but we'll use ``solarposition.pyephem()`` because it returns both the true (not refraction-corrected) and the apparent (refraction corrected) solar position. The airmass functions require the true zenith or apparent zenith at sea level, depending on the model. Calculating the relative airmass using a non-zero-altitude apparent zenith will result in incorrect results.

````python
ephem_data = pvlib.solarposition.pyephem(times, tus)
````

Calculate and plot the relative airmass. The default model is Young (1994), which requires the true zenith.

````python
pvlib.airmass.relativeairmass(ephem_data['zenith']).plot()
plt.ylabel('airmass')
````

Note that the airmass is defined as
```
AM = NaN if zenith > 90 else AM
```

A comparison of a few models over the course of a day.

````python
pvlib.airmass.relativeairmass(ephem_data['apparent_zenith']).plot(label='kastenyoung1989') # dangerous since we calculated airmass at a non-zero altitude.
pvlib.airmass.relativeairmass(ephem_data['zenith'], model='young1994').plot(label='young1994')
pvlib.airmass.relativeairmass(ephem_data['zenith'], model='simple').plot(label='simple')
    
plt.legend()
plt.ylim(0,50)
plt.ylabel('airmass')
````

A plot of air mass vs. zenith angle. See Reno (2012), Figure 5.

````python
simple = pvlib.airmass.relativeairmass(ephem_data['zenith'], model='simple')
plt.plot(ephem_data['zenith'], simple, label='simple')

young1994 = pvlib.airmass.relativeairmass(ephem_data['zenith'], model='young1994')
plt.plot(ephem_data['zenith'], young1994, label='young1994')

kastenyoung1989 = pvlib.airmass.relativeairmass(ephem_data['zenith'], model='kastenyoung1989')
plt.plot(ephem_data['zenith'], kastenyoung1989, label='kastenyoung1989')

plt.xlim(75, 90)
plt.xlabel('Zenith Angle (Degrees)')
plt.ylim(0,40)
plt.ylabel('Air Mass')

plt.legend(loc='upper left')
````

Next, we convert the relative air mass to absolute air mass. The relationship between the two is:
```
absolute airmass = (relative airmass) * pressure / 101325
```
where ``pressure`` is in Pascal.

````python
pvlib.airmass.absoluteairmass(young1994).plot(label='101325')
pvlib.airmass.absoluteairmass(young1994, pressure=70000).plot(label='70000')
location_pressure = pvlib.pvl_alt2pres(tus.altitude)
print('The default pressure for {} is {} Pascal'.format(tus.name, location_pressure))
young1994_abs = pvlib.airmass.absoluteairmass(young1994, pressure=location_pressure)
young1994_abs.plot(label='{:.0f}'.format(location_pressure))
plt.legend(title='pressure')
plt.ylabel('airmass')
````

## Clear sky models

pvlib makes it easy to calculate clear sky expectations. All you need is a pandas.DatetimeIndex and a pvlib.Location!

````python
cs_ineichen = pvlib.clearsky.ineichen(times, tus)
cs_ineichen.plot()
````

But the clear sky functions also have a number of options to tweak the calculation.

The default Linke turbidity is found from a built-in lookup table. You can override this with either a number or a series.

````python
cs_ineichen = pvlib.clearsky.ineichen(times, tus, linke_turbidity=4)
cs_ineichen.plot()
````

If you're going to use the clear sky function repeatedly for the same time periods, you may find it advantageous to precalculate the solar position, air mass, and Linke turbidity.

````python
cs_ineichen = pvlib.clearsky.ineichen(times, tus, zenith_data=ephem_data['apparent_zenith'], 
                                      airmass_data=young1994_abs, linke_turbidity=4)
cs_ineichen.plot()
````
