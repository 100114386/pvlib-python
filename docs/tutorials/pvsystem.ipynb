```python
ashraeiam = pvsystem.ashraeiam(.05, thetas)
ashraeiam.plot()
plt.ylabel('ASHRAE modifier')
plt.xlabel('input index')

plt.figure()
plt.plot(thetas, ashraeiam)
plt.ylabel('ASHRAE modifier')
plt.xlabel('input angle (deg)')
```
